```python
import pandas as pd
```

```python
vds_spot = pd.read_csv('VDS_37_03_01_567003.txt', engine='python', header = None)
vds_spot.columns = ["집계일자", "집계시분", "VDS_ID", "차로유형구분코드", "교통량", "None"]
vds_spot = vds_spot.drop('None', axis=1)
```

```python
#vds 기준 데이터
vds_std = pd.read_csv('ETC_79_04_01_983031.txt', engine = 'python')
```

```python
vds_std.head()
```

```python
#vds 지점 데이터와 vds 기준 데이터 합치기
vds_spot_std = pd.merge(vds_spot,vds_std)
```

```python
#콘존 데이터 불러오기
cz_data = pd.read_csv('ETC_78_04_01_275032.txt', engine = 'python')
```

```python
#용인-양지 콘존 데이터 확인
cz_area_data = cz_data[cz_data.콘존명 == '용인IC-양지IC']
cz_area_data
```

```python
spot_std_area= vds_spot_std[vds_spot_std.콘존ID == '0500CZE140']
spot_std_area.head()
```

```python
#1시간 단위, 구간교통량 vds데이터 불러오기
vds_section = pd.read_csv('VDS_38_03_01_457869.txt', engine='python',header=None)
vds_section.columns = ["집계일자", "집계시분", "콘존ID", "차로유형구분코드", "교통량", "None"]
vds_section = vds_section.drop('None', axis=1)
vds_section.head()
```

```python
#vds 구간 교통량과 cz데이터 합치기
vds_sec_cz = pd.merge(vds_section, cz_data)
```

```python
#대상 구간 vds 구간 교통량 확인 
sec_cz_area = vds_sec_cz[vds_sec_cz.콘존ID == '0500CZE140' ]
```

```python
sec_cz_area.head()
```

```python
writer = pd.ExcelWriter('vds_data.xlsx')
vds_spot.to_excel(writer, 'vds_spot')
vds_std.to_excel(writer, 'vds_std')
vds_spot_std.to_excel(writer, 'vds_spot_std')
spot_std_area.to_excel(writer, 'area_vds_spot')
vds_section.to_excel(writer, 'area_vds_section')
writer.save()
```
